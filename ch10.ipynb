```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
'key2' : ['one', 'two', 'one', 'two', 'one'],
'data1' : np.random.randn(5),
'data2' : np.random.randn(5)})
df
```

```python
grouped = df['data1'].groupby(df['key1'])
grouped.mean()
```

```python
means = df['data1'].groupby([df['key1'], df['key2']]).mean()
means.unstack()
```

```python
states = np.array(['Ohio', 'California', 'California', 'Ohio',
'Ohio'])
years = np.array([2005, 2005, 2006, 2005, 2006])
df['data1'].groupby([states, years]).mean()
```

```python
columns = pd.MultiIndex.from_arrays([['US', 'US', 'US', 'JP',
'JP'],[1, 3, 5, 1, 3]],names=['cty', 'tenor'])
hier_df = pd.DataFrame(np.random.randn(4, 5), columns=columns)
hier_df
```

```python
tips = pd.read_csv('E:/jupyter-notebook/pydata-book-2nd-edition/examples/tips.csv')
print(tips)
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips[:6]
```

```python
grouped = tips.groupby([ 'smoker','day'])
grouped_pct = grouped['tip_pct']
grouped_pct.agg('mean')
```

```python
result = tips.groupby('smoker')['tip_pct'].describe()
result
```

```python
result.unstack('smoker')
```

```python
f = lambda x: x.describe()
grouped.apply(f)
```

```python
def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[-n:]
tips.groupby('smoker', group_keys=False).apply(top)
```

```python
tips.groupby(['smoker', 'day']).apply(top, n=1,column='total_bill')
```

```python
states = ['Ohio', 'New York', 'Vermont', 'Florida',
'Oregon', 'Nevada', 'California', 'Idaho']
group_key = ['East'] * 4 + ['West'] * 4
data = pd.Series(np.random.randn(8), index=states)
fill_mean = lambda g: g.fillna(g.mean())
data.groupby(group_key).apply(fill_mean)
```

```python
suits = ['H', 'S', 'C', 'D']
card_val = (list(range(1, 11)) + [10] * 3) * 4
base_names = ['A'] + list(range(2, 11)) + ['J', 'K', 'Q']
cards = []
for suit in ['H', 'S', 'C', 'D']:
    cards.extend(str(num) + suit for num in base_names)
deck = pd.Series(card_val, index=cards)
deck[:13]
```

```python
df = pd.DataFrame({'category': ['a', 'a', 'a', 'a', 'b', 'b', 'b', 'b'],
'data': np.random.randn(8),'weights': np.random.rand(8)})
df
```

```python
grouped = df.groupby('category')
get_wavg = lambda g: np.average(g['data'],
weights=g['weights'])
grouped.apply(get_wavg)
```
